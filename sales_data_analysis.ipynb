# Sales Data Analysis Project

## Objective
Analyze sales data to identify trends, patterns, and business insights.

```python
import pandas as pd
```

## Data Loading


```python
df = pd.read_csv("train.csv")
df.head()
```

## Data Understanding

```python
df.info()
df.describe()
# df.isnull().sum()
```

## Data Cleaning


```python
df.drop_duplicates(inplace=True)
df.head()
```

```python
# df.fillna(obj.ffill(), inplace=True)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

## Feature Engineering
Extracted Year, Month for time-based analysis.

```python
df['Year'] = df['Order Date'].dt.year 
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.month_name()
df['Day'] = df['Order Date'].dt.day
```

```python
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'],dayfirst=True)
df.info()
```

## Exploratory Data Analysis

#### Sales by Region

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

Insight:
- West region generates the highest overall sales
- East region also contributes significantly to total revenue
- South region shows comparatively lower performance, indicating potential growth opportunity

#### Sales by Category

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

Insight:
- Technology category contributes the highest sales among all categories
- Furniture and Office Supplies follow but with comparatively lower revenue
- Technology products appear to be the main revenue drivers

#### Monthly Sales Trend

```python
monthly_sales = df.groupby('Month Name')['Sales'].sum()
```

```python
monthly_sales = df.resample('ME', on='Order Date')['Sales'].sum()
```

Insight:
- Sales show a clear upward trend towards the end of the year
- Peak sales are observed in November and December, indicating strong year-end demand (possibly due to holidays and discounts)
- Lower sales are observed in early months like January and February

#### Top Products

```python
top_products = df.groupby('Product Name')["Sales"].sum().sort_values(ascending=False)
top_products
```

Insight:
- A small number of products contribute disproportionately to total sales
- High-value products dominate the top positions
- This indicates a concentration of revenue among key products

### Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title("Sales by Region")
plt.show()
```

#### Monthly Trend

```python
monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.show()
```

```python
region_sales.plot(kind='bar', title="Sales by Region")
plt.show()
```

```python
category_sales.plot(kind='bar', title="Sales by Category")
plt.show()
```

## Final Insights

- Sales are highest in the West region, followed by the East region
- Technology category is the most profitable and dominant
- Sales peak during year-end months (November–December)
- Revenue is concentrated among a few top-performing products

## Conclusion

This analysis highlights key business trends such as regional performance, seasonal demand, and product contribution. These insights can help businesses make data-driven decisions, optimize inventory, and focus on high-performing areas.
